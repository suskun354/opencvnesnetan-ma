```python
import cv2
import numpy as np
import os

# Paths to YOLO model files
model_dir = 'C:/Users/Yunus Emre Duran/Documents/Python Çalışmalar/Canlı Nesne Tanıma'
cfg_path = os.path.join(model_dir, 'yolov3.cfg')
weights_path = os.path.join(model_dir, 'yolov3.weights')
names_path = os.path.join(model_dir, 'coco.names')

# Ensure the files are accessible
assert os.path.exists(cfg_path), f"Config file not found at {cfg_path}"
assert os.path.exists(weights_path), f"Weights file not found at {weights_path}"
assert os.path.exists(names_path), f"Names file not found at {names_path}"

# Load YOLO model
net = cv2.dnn.readNet(weights_path, cfg_path)
net.setPreferableBackend(cv2.dnn.DNN_BACKEND_OPENCV)
net.setPreferableTarget(cv2.dnn.DNN_TARGET_CPU)

# Load class names
with open(names_path, 'r') as f:
    classes = f.read().strip().split('\n')

# Initialize webcam
webcam = cv2.VideoCapture(0)

while True:
    ret, frame = webcam.read()
    if ret:
        # Prepare the frame for the neural network
        blob = cv2.dnn.blobFromImage(frame, 1/255.0, (416, 416), swapRB=True, crop=False)
        net.setInput(blob)
        
        # Get the names of all layers in the network
        layer_names = net.getLayerNames()
        output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers()]

        # Perform forward pass
        detections = net.forward(output_layers)
        
        # Initialize lists to hold the detected bounding boxes, confidences, and class IDs
        boxes = []
        confidences = []
        class_ids = []

        # Loop over each detection
        for output in detections:
            for detection in output:
                scores = detection[5:]
                class_id = np.argmax(scores)
                confidence = scores[class_id]
                
                if confidence > 0.5:
                    box = detection[0:4] * np.array([frame.shape[1], frame.shape[0], frame.shape[1], frame.shape[0]])
                    (centerX, centerY, width, height) = box.astype("int")
                    x = int(centerX - (width / 2))
                    y = int(centerY - (height / 2))
                    
                    boxes.append([x, y, int(width), int(height)])
                    confidences.append(float(confidence))
                    class_ids.append(class_id)
        
        # Apply non-maxima suppression to suppress weak, overlapping bounding boxes
        indices = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)

        # Draw bounding boxes
        if len(indices) > 0:
            for i in indices.flatten():
                (x, y) = (boxes[i][0], boxes[i][1])
                (w, h) = (boxes[i][2], boxes[i][3])

                color = [int(c) for c in np.random.randint(0, 255, size=(3,))]
                cv2.rectangle(frame, (x, y), (x + w, y + h), color, 2)
                text = f"{classes[class_ids[i]]}: {confidences[i]:.2f}"
                cv2.putText(frame, text, (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
        
        # Display the output
        cv2.imshow('Real-time Object Detection', frame)
        
        # Exit on pressing 'q'
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

# Release the webcam and close windows
webcam.release()
cv2.destroyAllWindows()
```
